## 확률(Probability)
- 주어진 `확률 분포가 고정`된 상태에서, `관측되는 사건이 변화`될 때, 확률을 표현하는 단어입니다.
- Probability는 특정 사건이 일어날 확률

$$
확률 = P(관측값 X | 확률분포 D)
$$

몇가지 예를 들면 금방 이해하실 것 같습니다.
- 동전을 던졌을 때 앞면이 나올 확률은 1/2이다.
- 주사위를 던졌을 때 3이 나올 확률은 1/6이다.

image.png

## 가능도(Likelihood)
- `관측된 사건이 고정`된 상태에서, `확률 분포가 변화`될 때(=확률 분포를 모를 때 = 가정할 때), 확률을 표현하는 단어입니다.
- Likelihood는 그 사건들을 설명할 수 있는 통계 모델

$$
가능도 = L(확률분포 D | 관측값 X)
$$

# [최대가능도 추정량(Maximum Likelihood Estimator, MLE)](https://jjangjjong.tistory.com/41)
- `최대 우도 추정`이란 각 관측값에 대한 총 가능도(모든 가능도의 곱)가 최대가 되게하는 분포를 찾는 것이라고 할 수 있다.

첫번째 그림을 보자.
- 우리는 가능도를 성명하기 위해 관측값이 하나일 경우를 말했지만, 아래처럼 여러 개의 관측값을 구했다고 가정해보자.
- 이때 `이렇게 관측 될 가능성이 가장 큰 확률분포는 무엇일까?`를 풀어내는 것이 Maximum likelihood 이다.

- 즉 우리가 수집한 관측값들이 나올 수 있는 가장 가능한 확률분포는 가능도가 제일큰, 검은 점이 제일 높이 위치한 정규분포에서 왔다고 추정하면 된다.

## [이진분류 - Sigmoid](https://www.youtube.com/watch?v=egeOjGEG1j4)

- 165cm인 사람이 있다. 이 사람이 남자일 확률은?
  - 무작위로 165cm인 사람 100명을 모집했을 때, 남자는 40명, 여자는 60명이라면,
  - 남자 165cm일 확률은 = 40/100 = 0.4일 것이다.

- 수식: P(남자 | 키가 165cm) = 1 - P(여자 | 키가 165cm)
- 이진분류일때는 로지스틱 회귀모델을 사용하며, 아래 수식은 로지스틱 회귀모델이다.
$$
P(남자|165cm) = { 1 \over 1 + e^{-b-\lambda f}} = { 1 \over 1 + e^{-b-\lambda 165}}
$$

아래는 대한민국 성인 키 데이터를 Gaussian(표준)분포로 변환한 그래프이다.
- 즉, 모델이 학습을 통해서 데이터를 이용해서 `최대가능도 추정량`에 어울리는 확률밀도함수를 만들 수 있다면,  
- 아래와 같이 여성에 대한 확률밀도함수를 구하고, 남자에 대한 확률밀도함수를 구할 수 있게 된다면, 165cm의 사람에 대한 예측을 확률적으로 할 수 있게 되는 것이다.





image.png

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 +np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## [다중분류 SoftMax](https://towardsdatascience.com/sigmoid-and-softmax-functions-in-5-minutes-f516c80ea1f9)

- 각 클래스에 대한 확률을 출력
- 입력받은 값들을 0~1사이의 값들로 모두 정규화하며 출력값들의 합은 항상 1이 되는 특성을 가진 함수

$$
softmax(z) = \frac{e^{z_{i}}}{\sum_{j=1}^K e^{z_{j}}} \ \ \ for\ i=1,2,\dots,K
$$

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

x = np.array([1.0,1.0,2.0])

y = softmax(x)

# y의 모든 확률 값을 합하면, 1(100%)임!!
print(np.sum(y))

ratio = y
labels = y

plt.pie(ratio, labels=labels, shadow=True, startangle=90)
plt.show()
```

# [다중 분류 예제](https://cypision.github.io/machine-learning/Imbalanced_Precision_Recall_Curve/)

## Load Data

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
# 데이터셋
digits = load_digits()

# features: target을 예측하기 위해서 필요한 데이터
data = digits.data
# target: 우리가 알고 싶은 데이터
target = digits.target

SEED = 42
# 모델을 학습결과를 평가하기 위해서 사전에 테스트 데이터 분리
x_train,x_valid,y_train,y_valid = train_test_split(data,target,random_state=SEED)

x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

# 벡터 변환
digit_img = x_valid[0].reshape(8, 8)

plt.imshow(digit_img, cmap="gray")
plt.axis("off")
plt.show()
```

## Modeling

```python
from sklearn.tree import DecisionTreeClassifier
# 모델 정의: max_depth -> 하이퍼파라미터
tree = DecisionTreeClassifier(max_depth = 5 , random_state=SEED)
# 모델 학습: features & target -> 지도학습
tree.fit(x_train,y_train)
# 모델 예측: predict() -> 정수 & predict_proba() -> 확률
pred = tree.predict(x_valid)
```

## 평가

### F1-Score
- micro
> 전체 클래스의 대하여 TP/FP/FN 구한 뒤에 F1-Score 계산
- macro
> 각 클래스에 대하여 F1-Score 계산뒤 산술 평균
- weighted
> 각 클래스에 대하여 F1-Score 계산뒤 각 클래스가 차지하는 비율에 따라 가중평균

```python
from sklearn.metrics import f1_score

f1_score(y_valid, pred, average="micro")
```

```python
f1_score(y_valid, pred, average="macro")
```

```python
f1_score(y_valid, pred, average="weighted")
```
